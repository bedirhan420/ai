```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("C:/Users/bedir/Downloads/odev_tenis.csv")
data
```

```python
outlook=data.iloc[:,0:1].values
le=preprocessing.LabelEncoder()
outlook[:,0]=le.fit_transform(data.iloc[:,0])
ohe=preprocessing.OneHotEncoder()
outlook=ohe.fit_transform(outlook).toarray()
outlook
```

```python
p=data.iloc[:,-1:].values
p[:,-1]=le.fit_transform(data.iloc[:,-1])
p=ohe.fit_transform(p).toarray()
p
```

```python
w=data.iloc[:,3].values.reshape(-1,1)
w=ohe.fit_transform(w).toarray()
w
```

```python
numeric=data[["temperature","humidity"]]
numeric
```

```python
sonuc=pd.DataFrame(data=outlook[:, :2],index=range(14),columns=["overcast","rainy"])
sonuc2=pd.DataFrame(data=w[:,:1],index=range(14),columns=["windy"])
sonuc3=pd.DataFrame(data=p[:,:1],index=range(14),columns=["play"])

```

```python
s=pd.concat([sonuc,numeric,sonuc2],axis=1)
s
```

```python
veri=pd.concat([s,sonuc3],axis=1)
veri
```

```python
x_train,x_test,y_train,y_test=train_test_split(s,sonuc3,test_size=0.3,random_state=0)
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)

y_pred=regressor.predict(x_test)
y_pred
```
